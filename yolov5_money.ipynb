```python
import zipfile
import os

zip_file_path = "/content/drive/MyDrive/pic.zip"

# Specify the directory to extract the zip file contents
extract_directory = "/content/drive/MyDrive/"

# Create the directory to extract the zip file contents
os.makedirs(extract_directory, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_directory)
```

```python
import glob

# Specify the path to the extracted image folder
image_folder_path = "/content/drive/MyDrive/100"

# Create a list of image file paths
image_paths = glob.glob(os.path.join(image_folder_path, "*.jpg"))
```

```python
from PIL import Image
import os

# Create a directory to save cropped images
output_dir = "/content/drive/MyDrive/100/cropped"
os.makedirs(output_dir, exist_ok=True)

# Loop over each image
for image_path in image_paths:
    # Open the image
    image = Image.open(image_path)

    # Run object detection on the image using YOLOv5
    # Replace this line with your own YOLOv5 detection code
    # bboxes = run_object_detection(image)
    x1 = 52
    y1 = 149
    x2 = 579
    y2 = 404

    # Example bounding box coordinates
    bboxes = [[x1, y1, x2, y2]]  # List of bounding box coordinates

    # Desired crop size
    crop_width = 526
    crop_height = 252

    # Crop and save each object based on the bounding box coordinates
    for i, bbox in enumerate(bboxes):
        # Extract the coordinates
        x1, y1, x2, y2 = bbox

        # Calculate the crop region
        crop_box = (x1, y1, x1 + crop_width, y1 + crop_height)

        # Crop the image
        cropped_image = image.crop(crop_box)

        # Save the cropped image with a unique name
        filename = os.path.basename(image_path)
        cropped_image_path = os.path.join(output_dir, f"{filename}_100_{i}.jpg")
        cropped_image.save(cropped_image_path)
```
